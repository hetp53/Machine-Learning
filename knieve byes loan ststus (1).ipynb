```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df["Default Status"]=encoder.fit_transform(df["Default Status"])
```

```python
df.head()
```

```python
x=df.drop(["Default Status","ID"],axis=1)
y=df["Default Status"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=90)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
yp=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,classification_report,precision_score,confusion_matrix
```

```python
accuracy_score(y_test,yp)
```

```python
recall_score(y_test,yp)
```

```python
print(classification_report(y_test,yp))
```

```python
precision_score(y_test,yp)
```

```python
confusion_matrix(y_test,yp)
```
