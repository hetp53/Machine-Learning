## outlier detectioon and removal

```python
# 1 z-score method- normally distributed data
# 2 3 STd method- skewed data
# 3 IQR (inter quartile range)- skewed data
```

### 1. z-score method

```python
df=[29,35,67,23,45,22,54,33,12,66,58,37,99,300,400,97,79,37,42]
```

```python
import pandas as pd
```

```python

data=pd.DataFrame(df,columns=["values"])
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
data.hist()
```

```python
 # z-score= (x-mean)/std
```

```python
# range= -3 to 3
```

```python
mean=data["values"].mean()
```

```python
std=data["values"].std()
```

```python
# it is not an outlier because ans is between -3 to 3
(66-mean)/std
```

```python
# it is outlier because ans is not between -3 to 3
(400-mean)/std
```

```python
# to find z score of all values
z_score=(data["values"]-mean)/std
```

```python
z_score
```

```python
data["z_score_value"]=z_score
```

```python
data
```

```python
data[(data["z_score_value"] < -3) | (data["z_score_value"] > 3)]
```

### 3 STd method- skewed data

```python
# lower bound= mean + (-3 std)
# upper bound= mean + (+3 std)
```

```python
df=[29,35,67,23,45,22,54,33,12,66,58,37,99,300,400,97,79,37,42]
data=pd.DataFrame(df,columns=["values"])
```

```python
data
```

```python
mean
```

```python
std
```

```python
lower_bound= mean + (-3 * std)
lower_bound
```

```python
upper_bound= mean + (+3 * std)
upper_bound
```

```python
data[(data["values"]< lower_bound) | (data["values"]> upper_bound)]
```

### IQR (inter quartile range)

```python
# IQR=q3-q1 = q2(50%) (median)
# LB=q1 - 1.5*IQR
# UB=q3 + 1.5*IQR
```

```python
df=[29,35,67,23,45,22,54,33,12,66,58,37,99,300,400,97,79,37,42]
data=pd.DataFrame(df,columns=["values"])
```

```python
data
```

```python
data["values"]=sorted(data["values"])
```

```python
data
```

```python
import numpy as np
```

```python
q1,q3=np.percentile(data["values"],[25,75])
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1 
IQR
```

```python
LB=q1 - 1.5*IQR
LB
```

```python
UB=q3 + 1.5*IQR
UB
```

```python
# outlier detection
data[(data["values"]< LB) | (data["values"]> UB)]
```

```python
# outlier removal
data[(data["values"]> LB) & (data["values"]< UB)]
```
